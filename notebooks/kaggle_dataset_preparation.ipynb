```python
import numpy as np
import pandas as pd
from tqdm import tqdm
from collections import defaultdict

from sklearn.preprocessing import KBinsDiscretizer
```

```python
transactions = pd.read_csv('../data/kaggle_transactions_data/transactions.csv')
tr_types = pd.read_csv('../data/kaggle_transactions_data/tr_types.csv', sep=';')
gender_train = pd.read_csv('../data/kaggle_transactions_data/gender_train.csv')
tr_mcc_codes = pd.read_csv('../data/kaggle_transactions_data/tr_mcc_codes.csv', sep=';')
```

```python
transactions.head()
```

```python
transactions.shape
```

```python
transactions_train = transactions[transactions.customer_id.isin(gender_train.customer_id)]
```

```python
transactions_train.shape
```

```python
transactions_train.head()
```

```python
encoders = {}
n_bins = 10

transactions_train['amount_bin'] = 0
for group in transactions_train.mcc_code.unique():
    encoders[group] = KBinsDiscretizer(n_bins=n_bins, encode='ordinal')
    condition = (transactions_train.mcc_code == group)
    values = transactions_train.loc[condition, 'amount'].values.reshape(-1, 1)
    
    transactions_train.loc[condition, 'amount_bin'] = encoders[group].fit_transform(values).astype(np.int32)
```

```python
transactions_train.head()
```

```python
transactions_train['trans_id'] = transactions_train['amount_bin'].astype(str) + \
    '_' + transactions_train['mcc_code'].astype(str) + \
    '_' + transactions_train['tr_type'].astype(str)
```

```python
transactions_train.head()
```

```python
transactions_train.trans_id.nunique()
```

```python
train_data = dict()
trans_encoder = defaultdict(set)

for idx, row in tqdm(transactions_train.iterrows()):
    client_id = row['customer_id']
    if client_id not in train_data:
        train_data[client_id] = defaultdict(list)
    
    trans_date = row['tr_datetime'].split(' ')[0]
    if trans_date not in trans_encoder[client_id]:
        trans_encoder[client_id].add(trans_date)

    trans_date_encoded = len(trans_encoder[client_id]) - 1
    train_data[client_id][trans_date_encoded].append(row['trans_id'])
```

```python
import pickle
pickle.dump(train_data, open('../data/kaggle_transactions_data/train_data_transactions.pkl', 'wb'))
```

```python
client2gender = dict(gender_train.values)
```

```python
train_df = []
num_days = 5

for client_id, transactions in tqdm(train_data.items()):
    examples = defaultdict(list)
    for day, ids in transactions.items():
        day_id = int(day / num_days)
        examples[day_id].extend(ids)
    
    for _, ex in examples.items():
        train_df.append([client_id, ' '.join(ex), client2gender[client_id]])
        
```

```python
final_dataset = pd.DataFrame(train_df, columns=['client_id', 'transactions', 'label'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
tr, te = train_test_split(final_dataset[['transactions', 'label']], random_state=24, test_size=0.05)
```

```python
tr.shape, te.shape
```

```python
tr_lens = tr.transactions.apply(lambda x: len(x.split())).values
te_lens = te.transactions.apply(lambda x: len(x.split())).values
```

```python
min_len = 2
max_len = 20
```

```python
tr = tr[(tr_lens >= min_len) & (tr_lens <= max_len)]
te = te[(te_lens >= min_len) & (te_lens <= max_len)]
```

```python
from collections import Counter
transactions = Counter()
```

```python
_ = [transactions.update(trans.split()) for trans in tr.transactions]
```

```python
from allennlp.data.vocabulary import DEFAULT_OOV_TOKEN
```

```python
DEFAULT_OOV_TOKEN
```

```python
vocabulary = [tok for tok, count in transactions.most_common()]
vocabulary.insert(0, DEFAULT_OOV_TOKEN)
```

```python
vocabulary = {tok: f'id_{i}' for i, tok in enumerate(vocabulary)}
```

```python
len(vocabulary)
```

```python
tr.transactions = tr.transactions.apply(
    lambda x: ' '.join([vocabulary.get(tok, vocabulary[DEFAULT_OOV_TOKEN]) for tok in x.split()])
)
```

```python
te.transactions = te.transactions.apply(
    lambda x: ' '.join([vocabulary.get(tok, vocabulary[DEFAULT_OOV_TOKEN]) for tok in x.split()])
)
```
